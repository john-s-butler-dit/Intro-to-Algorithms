```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility


N = 5000
a_x=331*4-1
c_x=701
m_x=5507*4*13
a_y=1021*4-1
c_y=709
m_y=5507*4*11
X=np.zeros(N)
Y=np.zeros(N)
x=np.zeros(N)
y=np.zeros(N)
X[0]=3
Y[0]=7
Green=0
fig = plt.figure()   

for i in range(0, N-1):
    X[i+1]= (a_x*X[i]+c_x)%m_x
    Y[i+1]= (a_y*Y[i]+c_x)%m_y
    
    x[i+1]=X[i+1]/(m_x-1)
    y[i+1]=Y[i+1]/(m_y-1)
    r = (y[i+1]-(x[i+1]*x[i+1]-x[i+1]+1))
    if r>=0:
        plt.scatter(x[i+1], y[i+1], marker='^',c="r")
    else:
        plt.scatter(x[i+1], y[i+1],  marker='o',c="g")
        Green=Green+1
    plt.axis('equal')
# Show the boundary between the regions:
plt.plot(x, x*x-x+1,'ko')
plt.xlim(0,1)
plt.ylim(0,1)
area=Green/N
plt.title('Area= %.5f, N =%.0f'%(area,N))
plt.show()
```
